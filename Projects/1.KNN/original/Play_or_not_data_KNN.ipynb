```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
df=pd.read_csv('play_or_not.csv')
df
```

```python
#creating labelEncoder
le=preprocessing.LabelEncoder()

#converting string labels into numbers
weather_encoded=le.fit_transform(df['weather'])
print(weather_encoded)
```

```python
#converting string labels into numbers
temp_encoded=le.fit_transform(df['temp'])
print(temp_encoded)

label=le.fit_transform(df['play'])
print(label)

#combining weather and temp into single listof tuples
features=list(zip(weather_encoded, temp_encoded))
print(features)
```

Caveat: If you do not convert them to tuples or another array-like structure and try to pass just the two lists directly to an algorithm, the algorithm may not know how the elements in the two lists correspond to each other or how they form the individual instances or samples.

```python
#Let's build our KNN classifier model
from sklearn.neighbors import KNeighborsClassifier

model=KNeighborsClassifier(n_neighbors=3)

#Train the model using the training sets
model.fit(features,label)
```

```python
# Get user input with instructions
print("Enter weather conditions:")
print("0: Overcast")
print("1: Rainy")
print("2: Sunny")
weather_input = int(input("Enter weather value (0, 1, or 2): "))

print("\nEnter temperature:")
print("0: Cool")
print("1: Hot")
print("2: Mild")
temp_input = int(input("Enter temperature value (0, 1, or 2): "))

# Predict output
predicted = model.predict([[weather_input, temp_input]])

# Print prediction
if predicted[0] == 1:
    print("\nPrediction: Play")
else:
    print("\nPrediction: Don't Play")
```

We skip the precision, recall, and f1 score due to limited data size.
