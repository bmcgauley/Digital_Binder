```python
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Lab5/suv_purchase.csv')
df
```

```python
# Convert the 'gender' feature and 'purchased' target from string to numeric

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
df['Purchased'] = le.fit_transform(df['Purchased'])
df
```

```python
# Let's split the dataset between x and y where x is all features such as gender, age, and estimatedsalary and y is the target

x = df.iloc[:, [1, 2, 3]].values
y = df.iloc[:, 4].values
```

```python
#splitting the dataset into training and test set.
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
#feature Scaling
from sklearn.preprocessing import StandardScaler
st_x=StandardScaler()
x_train=st_x.fit_transform(x_train)
x_test=st_x.transform(x_test)
```

```python
#fitting logistic regression to the training set
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression(random_state=0)
classifier.fit(x_train,y_train)
```

```python
#predicting the test set result
y_pred=classifier.predict(x_test)
```

```python
# Determine the model accuracy

from sklearn.metrics import accuracy_score

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of the model: {accuracy}")
```

```python
# Evaluate the model using precision, recall, FI score

from sklearn.metrics import precision_score, recall_score, f1_score

# Calculate precision, recall, and F1-score
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-score: {f1}")
```

```python

#Let's enter new data


while True:
    try:
        gender = input("Enter Gender (0 for Female, 1 for Male): ")
        gender = int(gender)
        if gender not in [0, 1]:
            raise ValueError
        age = int(input("Enter Age: "))
        estimated_salary = int(input("Enter Estimated Salary: "))

        new_data_point = nm.array([[gender, age, estimated_salary]])
        new_data_point = st_x.transform(new_data_point)
        prediction = classifier.predict(new_data_point)

        if prediction[0] == 0:
          print("Prediction: Not Purchased")
        else:
          print("Prediction: Purchased")

        break  # Exit the loop if input is valid
    except ValueError:
        print("Invalid input. Please enter valid numeric values for gender (0 or 1), age and estimated salary.")
    except Exception as e:
        print(f"An error occurred: {e}")
```
