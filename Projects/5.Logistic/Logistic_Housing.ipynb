```python
df = df.drop(columns=['address'])
```

```python
# prompt: eliminate unnamed feature

df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df
```

```python
# eliminate the data where null value in sqft feature
df = df.dropna(subset=['sqft'])
df
```

```python
# prompt: Turn neighborhood and status features, string to numeric using labelencoder

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

# Fit and transform the 'neighborhood' column
df['neighborhood'] = le.fit_transform(df['neighborhood'])

# Fit and transform the 'status' column
df['status'] = le.fit_transform(df['status'])

df
```

```python
# Because we only want the sold homes to account for this model, we eliminate the listed homes. 1 is sold and 0 is listed.
# Fit and transform the 'status' column
df['status'] = le.fit_transform(df['status'])

df = df[df['status'] != 0]
df
```

```python
# Eliminate data that has 1000 or less in price
# This is to prevent any unrealistic entries (e.g. human data entry error)

df = df[df['price'] > 1000]
df
```

```python
# Let's split the dataset into x (bedrooms, sqft, neighboorhood and status) and y (price).

# Define features (x) and target (y)
x = df[['bedrooms', 'sqft', 'neighborhood', 'status']]
y = df['price']
```

```python
# Let's do a regression analysis

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Split data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Initialize and train a linear regression model
model = LinearRegression()
model.fit(x_train, y_train)

# Make predictions on the test set
y_pred = model.predict(x_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```
